## Income Poverty and Government Assistance

```python
import pandas as pd
import numpy as np
```

```python
df_income = pd.read_csv('../data_files/census_data/census_melt_files/house_income.csv', index_col = 0)
df_income
```

```python
df_income.describe()
```

```python
df_income_gb = df_income.sort_values(by = ['count_house_income'], ascending = False)
df_income_gb
```

```python
df_total_income = df_income.groupby('house_income').sum()
df_total_income.sort_values(by = 'house_income', ascending = False)
```

```python
df_total_income = df_total_income.apply(lambda x: x/x.sum())*100
df_total_income.sort_values(by = 'count_house_income', ascending = False)
```
