## Commute Method

```python
import pandas as pd
import numpy as np
```

```python
df_commute_method = pd.read_csv('../data_files/census_data/census_melt_files/commute_method.csv', index_col = 0)
```

```python
df_commute_method.head()
```

```python
df_commute_method.describe()
```

```python
df_commute_method_bg = df_commute_method.sort_values(by = ['count_commute_method'], ascending = False)
df_commute_method_bg
```

```python
df_commute_by_method = df_commute_method.groupby('commute_method').sum()
df_commute_by_method.sort_values(by = 'count_commute_method', ascending = False)
```

```python
df_commute_by_method = df_commute_by_method.apply(lambda x: x/x.sum())*100
df_commute_by_method.sort_values(by = 'count_commute_method', ascending = False)
```
