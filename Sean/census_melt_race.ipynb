## Race

```python
import pandas as pd
import numpy as np
```

```python
df_race = pd.read_csv('../data_files/census_data/census_melt_files/race.csv', index_col = 0)
df_race
```

```python
df_race.describe()
```

```python
df_race_gb = df_race.sort_values(by = ['count_race'], ascending = False)
df_race_gb
```

```python
df_total_race = df_race.groupby('race').sum()
df_total_race.sort_values(by = 'count_race', ascending = False)
```

```python
df_total_race = df_total_race.apply(lambda x: x/x.sum())*100
df_total_race.sort_values(by = 'count_race', ascending = False)
```
