## What Types of Companies Men and Women Work For

### Originating data split by binary gender. All employees who work are split into 5 work-type categories: for profit, government, not for profit, self employed and unpaid family caretaker (how is this last one defined?)

```python
import pandas as pd
import numpy as np
```

```python
df_biz_type_female = pd.read_csv('../data_files/census_data/census_melt_files/biz_type_female.csv', index_col = 0)
```

```python
df_biz_type_female.head()
```

```python
df_biz_type_female.describe()
```

```python
df_biz_type_female.sort_values(by = ['f_count_biz_type'], ascending = False)
```

```python
f_biz_setting = df_biz_type_female.groupby('f_biz_type').sum()
f_biz_setting.sort_values(by = 'f_count_biz_type', ascending = False)
```

```python
f_biz_setting_percent = f_biz_setting.apply(lambda x: x/x.sum())*100
f_biz_setting_percent.sort_values(by = 'f_count_biz_type', ascending = False)
```
