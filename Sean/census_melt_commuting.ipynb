## Commuting

## Data related to people commuting: commute method, commute time and time leaving for work.

```python
import pandas as pd
import numpy as np
```

### Commute method, Car SOV, Car HOV, WFH, Bus

```python
df_commute_method = pd.read_csv('../data_files/census_data/census_melt_files/commute_method.csv', index_col = 0)
```

```python
df_commute_method.head()
```

```python
df_commute_method.describe()
```

```python
df_commute_method_bg = df_commute_method.sort_values(by = ['count_commute_method'], ascending = False)
df_commute_method_bg
```

```python
df_commute_by_method = df_commute_method.groupby('commute_method').sum()
df_commute_by_method.sort_values(by = 'count_commute_method', ascending = False)
```

```python
df_commute_by_method = df_commute_by_method.apply(lambda x: x/x.sum())*100
df_commute_by_method.sort_values(by = 'count_commute_method', ascending = False)
```

### Commute time, bus & car

```python
df_commute_time_bus = pd.read_csv('../data_files/census_data/census_melt_files/commute_time_bus.csv', index_col = 0)
```

```python
df_commute_time_car = pd.read_csv('../data_files/census_data/census_melt_files/commute_time_car.csv', index_col = 0)
```

```python
df_commute_time_bus.head()
```

```python
df_commute_time_car.head()
```

```python
df_commute_time_bus.describe()
```

```python
df_commute_time_car.describe()
```

```python
df_commute_time_bus.sort_values(by = ['count_commute_time_bus'], ascending = False)
```

```python
df_commute_time_car.sort_values(by = ['count_commute_time_car'], ascending = False)
```

```python
df_bus_commute_time = df_commute_time_bus.groupby('commute_time_bus').sum()
df_bus_commute_time.sort_values(by = 'count_commute_time_bus', ascending = False)
```

```python
df_car_commute_time = df_commute_time_car.groupby('commute_time_car').sum()
df_car_commute_time.sort_values(by = 'count_commute_time_car', ascending = False)
```
