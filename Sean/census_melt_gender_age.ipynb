## Age and Gender

```python
import pandas as pd
import numpy as np
```

```python
df_gender = pd.read_csv('../data_files/census_data/census_melt_files/gender_census.csv', index_col = 0)
```

```python
df_gender.head()
```

```python
df_gender.describe()
```

```python
df_gender_gb = df_gender.sort_values(by = ['count_gender'], ascending = False)
df_gender_gb
```

```python
df_total_gender = df_gender.groupby('gender').sum()
df_total_gender.sort_values(by = 'count_gender', ascending = False)
```

```python
df_total_gender = df_total_gender.apply(lambda x: x/x.sum())*100
df_total_gender.sort_values(by = 'count_gender', ascending = False)
```

```python
df_gender_working_age = pd.read_csv('../data_files/census_data/census_melt_files/gender_working_age_census.csv', index_col = 0)
```

```python
df_gender_working_age
```

```python
df_gender_working_age.describe()
```

```python
df_gender_working_age_gb = df_gender_working_age.sort_values(by = ['count_gender_work_age'], ascending = False)
df_gender_working_age_gb
```

```python
df_total_gender_working_age = df_gender_working_age.groupby('gender_work_age').sum()
df_total_gender_working_age.sort_values(by = 'count_gender_work_age', ascending = False)
```

```python
df_total_gender_working_age = df_total_gender_working_age.apply(lambda x: x/x.sum())*100
df_total_gender_working_age.sort_values(by = 'count_gender_work_age', ascending = False)
```
